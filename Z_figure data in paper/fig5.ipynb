```python

def perform_tests(df):
    features = df.iloc[:, :-1]
    target = df.iloc[:, -1]
    for column in features:
        feature = features[column]    
        t_stat, t_p = stats.ttest_ind(feature, target)
        print(f"{column} t-test: statistic={t_stat}, p-value={t_p}")
    for column in features:
        feature = features[column] 
        u_stat, u_p = stats.mannwhitneyu(feature, target)
        print(f"{column} U-test: statistic={u_stat}, p-value={u_p}")

def perform_tests_1(df):
    features = df.iloc[:, :-2]
    target = df.iloc[:, -2]
    for column in features:
        feature = features[column]    
        t_stat, t_p = stats.ttest_ind(feature, target)
        print(f"{column} t-test: statistic={t_stat}, p-value={t_p}")
    for column in features:
        feature = features[column] 
        u_stat, u_p = stats.mannwhitneyu(feature, target)
        print(f"{column} U-test: statistic={u_stat}, p-value={u_p}")
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error

def calculate_mae(df1, df2):
    mae_list = []
    for column in df1.columns:
        mae = mean_absolute_error(df1[column], df2[column])
        mae_list.append(mae)
    return mae_list
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

def calculate_rmse(df1, df2):
    rmse_list = []
    for column in df1.columns:
        mse = mean_squared_error(df1[column], df2[column])
        rmse = np.sqrt(mse)
        rmse_list.append(rmse)
    return rmse_list
```

```python
def merge_csv_files(prefix):
    filenames = glob.glob(f'DATA/fig5/real_trna_data/{prefix}_*.csv')
    df_list = [pd.read_csv(filename) for filename in filenames]
    df = pd.concat(df_list, ignore_index=False)
    return df
df=merge_csv_files("real_tRNA_data")
tRNA_data=df.set_index("gene_id")

tissue_diff=pd.read_csv("DATA/fig5/select_differ_tissue.csv",index_col="tissue")
ana = pd.read_csv('DATA/fig5/sample_cancer_type.csv',index_col='Case.ID')
tRNA_data.columns = ana.loc[tRNA_data.columns, 'Primary.Site']
tRNA_data.columns = tissue_diff.loc[tRNA_data.columns, 'tissue_id']
select_list=tissue_diff[tissue_diff["type"]!=0]["tissue_id"].tolist()
select_data=tRNA_data[select_list]
```

```python
def plot_embedded(x,title):
    
    half = len(x) // 2
    colors = ['#FF6AC2'] * half + ['#64CCC5'] * (len(x) - half)
    plt.figure(figsize=(8, 8))
    scatter = plt.scatter(c_tRNA_data['UMAP_1'], 
                          c_tRNA_data['UMAP_2'],
                          s=5, 
                          c=colors,
                          marker=".", alpha=1)
    plt.title(title)
    plt.xlabel('UMAP Component 1')
    plt.ylabel('UMAP Component 2')
    plt.show()
```

```python
title_list=["all","LCV","COR","HF","rR","tR","HK","HCV","HF2","seened"]
```

```python
for title in title_list:
    df= select_data[select_list]
    tissue_name=df.iloc[429:df.shape[0], :].rank()
    data1=pd.DataFrame([])
    gene_set=title
    for i in range(1,38):
        data=pd.read_csv("DATA/fig5/pridict_tissue_"+gene_set+"/tissue_"+str(i)+"_pridict.csv",index_col=0).T
        data.columns=tissue_name["tissue_"+str(i)].columns
        data1=pd.concat([data1,data],axis=1)
    data1.index=select_data.iloc[0:429,1].index
    type_A=tissue_diff[tissue_diff["type"]==1]["tissue_id"].tolist()
    type_A=[item for item in type_A if item not in ['tissue_24', 'tissue_25']]
    type_B=tissue_diff[tissue_diff["type"]==2]["tissue_id"].tolist()
    type_A_data_pridict=data1.loc[:,type_A]
    type_B_data_pridict=data1.loc[:,type_B]
    type_A_data_pridict = type_A_data_pridict.sort_index(axis=1, ascending=True)
    type_B_data_pridict = type_B_data_pridict.sort_index(axis=1, ascending=True)
    type_A_data_real=tRNA_data.loc[:,type_A]
    type_B_data_real=tRNA_data.loc[:,type_B]
    type_A_data_real = type_A_data_real.sort_index(axis=1, ascending=True)
    type_B_data_real = type_B_data_real.sort_index(axis=1, ascending=True)
    c_tRNA_data = pd.concat([type_A_data_pridict.sample(n=2000, axis=1).T,type_A_data_real.sample(n=2000, axis=1).T],axis=0)
    reducer = umap.UMAP(n_components=2, random_state=40)
    c_tRNA_data_embedded = reducer.fit_transform(c_tRNA_data)
    c_tRNA_data['UMAP_1'] = c_tRNA_data_embedded[:, 0]
    c_tRNA_data['UMAP_2'] = c_tRNA_data_embedded[:, 1]
    x=c_tRNA_data['UMAP_1']
    plot_embedded(x,title)
```
