# fig3. a

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
explain_method=pd.read_csv("DATA/explain_method.csv",index_col="gene_id")
```

```python
explain_method=explain_method.iloc[:,[0,1,3]]
ig_rank=explain_method["ig_rank"].rank().tolist()
saliency_rank=explain_method["saliency_rank"].rank().tolist()
gs_rank=explain_method["gs_rank"].rank().tolist()


data = [(saliency_rank[i]+ig_rank[i]+gs_rank[i])/3 for i in range(len(saliency_rank))]
colors = np.array(data)
sorted_colors = np.sort(colors)
color_map = plt.cm.ScalarMappable(cmap='cool')
color_map.set_array(sorted_colors)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(saliency_rank, ig_rank, gs_rank, c=colors[::1], cmap='cool', s=2,alpha=0.5)

ax.xaxis.set_tick_params(direction='in')
ax.yaxis.set_tick_params(direction='in')
ax.zaxis.set_tick_params(direction='in')

ax.xaxis.set_label_coords(0.5, -0.08)
ax.yaxis.set_label_coords(-0.08, 0.5)
ax.zaxis.set_label_coords(-0.05, 0.5)

ax.set_xlabel('saliency rank', fontsize=12)
ax.set_ylabel('ig rank', fontsize=12)
ax.set_zlabel('gs rank', fontsize=12)
ax.grid(False)

plt.show()

```

# fig3. b

```python
correlation_matrix = np.corrcoef([saliency_rank, ig_rank, gs_rank])
pearson_correlation = correlation_matrix

print(pearson_correlation)
```

# fig3. c

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats


def mann_whitney_u_test(list1, list2):
    u_statistic, p_value = stats.mannwhitneyu(list1, list2)
    if p_value < 1e-20:
        p_value = 1e-20
    return u_statistic, p_value
```

```python
data1=pd.read_csv("DATA/fig3/ig_coeff_of_variation.csv")
data2=pd.read_csv("DATA/fig3/ig_coeff_of_variation_other.csv")
data3=pd.read_csv("DATA/fig3/gs_coeff_of_variation.csv")
data4=pd.read_csv("DATA/fig3/gs_coeff_of_variation_other.csv")
data5=pd.read_csv("DATA/fig3/saliency_coeff_of_variation.csv")
data6=pd.read_csv("DATA/fig3/saliency_coeff_of_variation_other.csv")
dataa=pd.DataFrame()

combined_data = [data1, data2, dataa,data3, data4,dataa,data5, data6]
plt.figure(figsize=(4, 6)) 
sns.boxplot(data=combined_data, width=0.7, notch=True, showfliers=False, color="#F1F0E8")
plt.ylabel("Coeff of Variation", fontsize=18)   
plt.xticks([0,1,3,4,6,7], ['Integrated Gradients', 'IG background','Gradient SHAP'
                              , 'GS background','Saliency', 'Saliency background'], fontsize=15,rotation=90) 
plt.yticks(fontsize=15)  
plt.tight_layout()
#plt.savefig("Coeff of Variation select.pdf",dpi=600)
plt.show()
```

```python
print(mann_whitney_u_test(data1, data2))
print(mann_whitney_u_test(data3, data4))
print(mann_whitney_u_test(data5, data6))
```

```python
data1=pd.read_csv("DATA/fig3/ig_mean_values.csv")
data2=pd.read_csv("DATA/fig3/ig_mean_values_other.csv")
data3=pd.read_csv("DATA/fig3/gs_mean_values.csv")
data4=pd.read_csv("DATA/fig3/gs_mean_values_other.csv")
data5=pd.read_csv("DATA/fig3/saliency_mean_values.csv")
data6=pd.read_csv("DATA/fig3/saliency_mean_values_other.csv")
dataa=pd.DataFrame()

combined_data = [data1, data2, dataa,data3, data4,dataa,data5, data6]
plt.figure(figsize=(4, 6)) 
sns.boxplot(data=combined_data, width=0.7, notch=True, showfliers=False, color="#ADC4CE")
plt.ylabel("Mean expression", fontsize=18)   
plt.xticks([0,1,3,4,6,7], ['Integrated Gradients', 'IG background','Gradient SHAP'
                              , 'GS background','Saliency', 'Saliency background'], fontsize=15,rotation=90) 
plt.yticks(fontsize=15)  
plt.tight_layout()
#plt.savefig("Mean expression select.pdf",dpi=600)
plt.show()
```

```python
print(mann_whitney_u_test(data1, data2))
print(mann_whitney_u_test(data3, data4))
print(mann_whitney_u_test(data5, data6))
```

```python
import os
import pandas as pd
import glob
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
def concatenate_split_files(directory, filename):
    base_filename, extension = os.path.splitext(filename)
    split_files = glob.glob(os.path.join(directory, f"{base_filename}_*{extension}"))
    data = pd.concat((pd.read_csv(file) for file in split_files))
    return data
tRNA_mRNA_corrmatrix=concatenate_split_files("DATA/fig3/tRNA_mRNA_corrmatrix_id","tRNA_mRNA_corrmatrix_id.csv").set_index("gene_id")
go_enrich_gene_set=pd.read_csv("DATA/fig3/go_enrich.csv",index_col="gene_id")

tRNA_mRNA_corrmatrix=tRNA_mRNA_corrmatrix.iloc[429:tRNA_mRNA_corrmatrix.shape[0],:]
mRNA_cor=pd.DataFrame(tRNA_mRNA_corrmatrix.abs().mean(axis=1))
go_enrich_gene_set=go_enrich_gene_set[go_enrich_gene_set.index.isin(tRNA_mRNA_corrmatrix.index)]
```

```python
#feature sum
df_mean=pd.DataFrame(explain_method.rank().mean(axis=1))
df_mean.index=df_mean.index.str.split('.', expand=True)
#plot_combinboxplot
combined_data=[random.sample(mRNA_cor[0].tolist(), 30)]
data = {}
for i in range(int(len(go_enrich_gene_set.columns.tolist())/2)):
    selected_list=go_enrich_gene_set.loc[go_enrich_gene_set[go_enrich_gene_set.columns.tolist()[i]]==1].index.tolist()
    selected=mRNA_cor.loc[selected_list][0].tolist()
    combined_data.append(selected)

plt.figure(figsize=(12, 6)) 
sns.boxplot(data=combined_data, width=0.4, notch=True, showfliers=False, color="#DFA67B")
plt.ylabel("Correlation coefficient with the tRNA level", fontsize=14)   
plt.xticks(list(range(1,38)),go_enrich_gene_set.columns.tolist()[0:37], fontsize=15,rotation=90) 
plt.yticks(fontsize=15)  
plt.tight_layout()
plt.show()

combined_data=[random.sample(mRNA_cor[0].tolist(), 30)]
data = {}
for i in range(int(len(go_enrich_gene_set.columns.tolist())/2),len(go_enrich_gene_set.columns.tolist())):
    selected_list=go_enrich_gene_set.loc[go_enrich_gene_set[go_enrich_gene_set.columns.tolist()[i]]==1].index.tolist()
    selected=mRNA_cor.loc[selected_list][0].tolist()
    combined_data.append(selected)

plt.figure(figsize=(12, 6)) 
sns.boxplot(data=combined_data, width=0.4, notch=True, showfliers=False, color="#DFA67B")
plt.ylabel("Correlation coefficient with the tRNA level", fontsize=14)   
plt.xticks(list(range(1,39)),go_enrich_gene_set.columns.tolist()[37:78], fontsize=15,rotation=90) 
plt.yticks(fontsize=15)  
plt.tight_layout()
#plt.savefig("correlation coefficient with the tRNA level 2.pdf",dpi=600)
plt.show()
```
