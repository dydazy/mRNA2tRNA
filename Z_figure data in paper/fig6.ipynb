# fig6 c

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('DATA/fig6/neuron_DE_other.csv')

grouped = df.groupby('group')['tRNA-Arg-TCT-4-1']

group_names = [name for name, _ in grouped]
group_data = [data for _, data in grouped]

plt.subplots(figsize=(8, 6))
box=plt.boxplot(group_data, labels=group_names,patch_artist=True,showfliers=False,widths=0.4)

plt.ylim(0,15000)
plt.title('DE_neruro_vs_other_tissue')
plt.xlabel('Group')
plt.ylabel('Value')

colors = ['#E8F6EF', '#FFB84C', 'lightgreen']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```

```python
df = pd.read_csv('DATA/fig6/neuron_DE_other.csv')

import pandas as pd
from scipy.stats import mannwhitneyu

def perform_mann_whitney_u_test(df):
    group1 = df[df["group"] == 1]["tRNA-Arg-TCT-4-1"]
    group2 = df[df["group"] == 2]["tRNA-Arg-TCT-4-1"]
    stat, p = mannwhitneyu(group1, group2)
    print((stat, p))
perform_mann_whitney_u_test(df)
```

# fig6 d-e

```python
import umap
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("DATA/fig6/ATAC_300_tRNA_neuron.csv",index_col="gene_id")
df=df.rank()
#df=np.log2(df+1)
# 创建UMAP对象
reducer = umap.UMAP(random_state=25)

# 拟合并转换数据
umap_data = reducer.fit_transform(df.T)

#tsen降维分析
plt.subplots(figsize=(6, 6))
plt.scatter(umap_data[0:28,0],  umap_data[0:28,1], marker=".", c='#F6BA6F', label='real',alpha=1,linewidths=5)
plt.scatter(umap_data[28:141,0],  umap_data[28:141,1], marker='.', c='#ADE4DB', label='pridict',alpha=1,linewidths=5)

plt.show()
```

```python
from sklearn.manifold import TSNE

df=pd.read_csv("DATA/fig6//ATAC_300_tRNA_neuron.csv",index_col="gene_id")
df=df.rank()
# 创建UMAP对象
tsne = TSNE(n_components=2, random_state=5)
X_tsne = tsne.fit_transform(df.T)
#tsen降维分析
plt.subplots(figsize=(6, 6))
plt.scatter(X_tsne[0:24,0],  X_tsne[0:24,1], marker=".", c='#F266AB', label='real',alpha=1,linewidths=5)
plt.scatter(X_tsne[24:141,0],  X_tsne[24:141,1], marker='.', c='#2CD3E1', label='pridict',alpha=1,linewidths=5)

plt.show()
```
